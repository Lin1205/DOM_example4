```python
df=pd.read_csv("MN.csv",dtype='unicode')
#get rid of rows with all data empty
df=df.dropna(axis=0,how='all')
```

```python
#convert some columns to integers
df=df[['Borough','SchoolDist','Council','ZipCode','PolicePrct','HealthArea','Address','ZoneDist1','ZoneDist2','ZoneDist3','ZoneDist4']]
```

```python
#get rid of rows with empty zipcode
df=df[df['ZipCode'].isnull() == False]
df.head(20)
```

```python
#get rid of rows with empty address
df=df[df['Address'].isnull() == False]
df.head(20)
```
